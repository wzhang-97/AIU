```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gx = np.fromfile('/home/wenzhang/multi unet/data/3rd/nx/0.dat', dtype=np.single).reshape((128,128,128))
gx = np.transpose(gx)
print(gx.shape)
```

```python
gx2_v1 = np.fromfile('/home/wenzhang/multi unet/data/3rd/gx_v/gx2_v1/0.dat', dtype=np.single).reshape((128,128,128))
gx2_v1 = np.transpose(gx2_v1)
print(gx2_v1.shape)
gx2_v2 = np.fromfile('/home/wenzhang/multi unet/data/3rd/gx_v/gx2_v2/0.dat', dtype=np.single).reshape((128,128,128))
gx2_v2 = np.transpose(gx2_v2)
print(gx2_v2.shape)
gx3_v1 = np.fromfile('/home/wenzhang/multi unet/data/3rd/gx_v/gx3_v1/0.dat', dtype=np.single).reshape((128,128,128))
gx3_v1 = np.transpose(gx3_v1)
print(gx3_v1.shape)
gx3_v2 = np.fromfile('/home/wenzhang/multi unet/data/3rd/gx_v/gx3_v2/0.dat', dtype=np.single).reshape((128,128,128))
gx3_v2 = np.transpose(gx3_v2)
print(gx3_v2.shape)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(gx[:,:,0])
plt.colorbar()
plt.show()
```

```python
gx2_tan = np.abs(np.arctan(gx2_v1 / gx2_v2)*180/np.pi)
gx3_tan = np.abs(np.arctan(gx3_v1 / gx3_v2)*180/np.pi)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(gx3_tan[:,:,0])
plt.colorbar()
plt.show()
```
